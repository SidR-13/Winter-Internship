```python
import pandas as pd
import sqlite3 as sql
```

```python
#connecting the database
con = sql.connect('/content/drive/MyDrive/Week 5/sql-murder-mystery.db')
```

```python
#looking for data from the crime scene report
query_1 = '''
SELECT *
FROM crime_scene_report
WHERE city = "SQL City"
ORDER BY date;
'''

#sewtting dataframe width to the max
pd.set_option('display.max_colwidth',None)

#running the query
pd.read_sql_query(query_1, con)
```

After some observation, we find that there are 2 witnesses.

```python
#checking the personal details of the witnesses
query_2 = '''
SELECT *
FROM person
WHERE address_street_name = "Northwestern Dr"
ORDER BY address_number desc;
'''

pd.read_sql_query(query_2, con)[:1]
```

```python
query_3 = '''
SELECT *
FROM person
WHERE name like '%Annabel%' AND address_street_name = "Franklin Ave";
'''
pd.read_sql_query(query_3, con)
```

```python
#Viewing the interview of the murder witnesses after the murder
query_4 = '''
SELECT *
FROM interview
WHERE person_id = 14887 OR person_id = 16371;
'''
pd.read_sql_query(query_4, con)
```

We get 2 major clues in the investigation-
The killer is a man and a member of the gym with the status being gold and having the membership no. starting as 48Z and left in a car with a plate including the initials as "H42W"

```python
#Checking the database with the above details
query_5 = '''
SELECT *
FROM get_fit_now_check_in
WHERE membership_id like "%48Z%" AND check_in_date = 20180109
ORDER BY check_in_date;
'''
pd.read_sql_query(query_5, con)
```

Two members found along with their membership id

```python
#Checking the details of the car
query_6 = '''
SELECT *
FROM drivers_license
WHERE plate_number like "%H42W%";
'''
pd.read_sql_query(query_6, con)
```

Two males found with the plate number containing the initials H42W

```python
#cheking the personal details of the two male
query_7 = '''
SELECT  *
FROM person
WHERE license_id = "423327" OR license_id = "664760";
'''
pd.read_sql_query(query_7, con)
```

```python
#To check which of the two males are a member of the gym
query_8 = '''
SELECT *
FROM get_fit_now_member
WHERE person_id = "51739" OR person_id = "67318";
'''
pd.read_sql_query(query_8, con)
```

Hence the killer has been found.

```python
query_10 = '''
SELECT *
FROM drivers_license
WHERE car_make = "Tesla" AND car_model = "Model S" AND gender = "female" AND hair_color = "red";
'''
pd.read_sql_query(query_10, con)
```
